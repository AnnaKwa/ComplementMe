```python
http = urllib3.PoolManager()

def get_links_uniqlo(query_string, num_images):
    # initialize place for links
    links = []
    # step by 100 because each return gives up to 100 links
    for i in range(0, num_images, 100):

        image_type="ActiOn"
        query= query_string.split()
        query='+'.join(query)
        
        url='https://www.uniqlo.com/us/en/search/?q='+query+'&lang=default'
  
        http = urllib3.PoolManager(cert_reqs='CERT_REQUIRED', ca_certs=certifi.where())
        
        response = http.request('GET', url)
        soup = BeautifulSoup(response.data.decode('utf-8'))
        html_content = soup.prettify()

        new_soup = Soup(html_content, 'lxml')

        # all img tags, only returns results of search
        imgs = new_soup.find_all('img')

        # loop through images and put src in links list
        for j in range(len(imgs)):
            if'prod' in imgs[j]["src"]:
                links.append(imgs[j]["src"])

    return links


def get_links_nordstrom(query_string, num_images, num_items_per_pg=200, sleep_time=1):
    # initialize place for links
    links = []
    # nordstrom shows set number of items per page
    n_pages = int((num_images-1)/num_items_per_pg + 1)
    
    f=open('nordstrom_40perpage.html','w')
    for n in range(n_pages):
        pg_num=n+1
        image_type="ActiOn"
        query= query_string.split()
        query='+'.join(query)
   
        url='https://shop.nordstrom.com/sr?origin=keywordsearch&keyword='+query+'&page='+str(pg_num)+'&top='+str(num_items_per_pg)
        print(url)
    
        http = urllib3.PoolManager(cert_reqs='CERT_REQUIRED', ca_certs=certifi.where())
        
        response = http.request('GET', url)
        
        time.sleep(sleep_time)
        soup = BeautifulSoup(response.data.decode('utf-8'))
        html_content = soup.prettify()
        
        print(html_content, file=f)

        new_soup = Soup(html_content, 'lxml')

        # all img tags, only returns results of search
        imgs = new_soup.find_all('img')
        print( len(imgs))
        # loop through images and put src in links list
        for j in range(len(imgs)):
            if'store/product' in imgs[j]["src"]:
                links.append(imgs[j]["src"])
            #links.append(imgs[j]["src"])
        f.close()
    return links
    
```
